```python
import re
def findVectorFromCSV(word):
    temp = re.sub(' +', ' ', word)
    temp_li = list(temp[1:-1].split())
    vector = []
    for st in temp_li:
        vector.append(float(st.rstrip("\n").rstrip(",")))
    return np.array(vector)
```
