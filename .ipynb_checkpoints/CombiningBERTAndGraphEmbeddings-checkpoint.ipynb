```python
import re
def findVectorFromCSV(word):
    temp = re.sub(' +', ' ', word)
    temp_li = list(temp[1:-1].split())
    vector = []
    for st in temp_li:
        vector.append(float(st.rstrip("\n").rstrip(",")))
    return np.array(vector)
```

```python
class sentenceToEmbedding:
    def __init__(self, sentence, embeddedSentence):
        self.sentence = sentence
        self.embeddedSentence = embeddedSentence
```

```python
def preProcessWords(word):
    temp = word.lower().rstrip(",").rstrip(".").rstrip(":").rstrip(";").rstrip("`")
    words = re.split('[- _ – : /]', temp) 
    return words
```

```python
comb_emb = pd.DataFrame(columns = ['word', 'embeddings'])
modified_dataset = pd.DataFrame(columns = ['text', 'keywords'])
```

```python
import gc
sent_emb = []
i = 0
j = 0
for index, row in data.iterrows():
    sentence = row['text']
    word_list = list(sentence.split(" "))
    modifiedSen = ""
    for word in word_list:
        w = preProcessWords(word)
        for item in w:
            try:
                v1 = findVectorFromCSV(bert_emb.loc[item]['BertEmbeddings'])       
                v2 = findVectorFromCSV(deepWalk_emb.loc[item]['DeepWalkEmbedding'])
                final_emb = np.add(v1, v2)
                comb_emb.loc[j] = [item] + [final_emb]
                j = j + 1
                modifiedSen = modifiedSen + item + " "
                del v1, v2
                gc.collect()
            except:
                print(item)
            
            del item
            gc.collect()
        del w
        gc.collect()
    
    modified_dataset.loc[i] = [modifiedSen] + [row['keys']]
    i = i + 1
    del modifiedSen
    gc.collect()
```

```python
comb_emb.to_csv("combinedEmbeddings.csv")
```

```python
modified_dataset.to_csv("ModifiedDataset.csv")
```
