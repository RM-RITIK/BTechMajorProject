```python
complete_tag_list=[]
def generate_BIO_tag(word_list , B_list , I_list):
    BIO_tagged_string=""
    for word in word_list:
        word=word.replace(',',"")
        if word in B_list:
            BIO_tagged_string+="B "
        elif word in I_list:
            print("Reachign here")
            if(BIO_tagged_string[len(BIO_tagged_string)-2:len(BIO_tagged_string)-1])=='B':
                BIO_tagged_string+="I "
        else:
            BIO_tagged_string+="O "
    BIO_tagged_string= BIO_tagged_string.lstrip()
    BIO_tagged_string= BIO_tagged_string.rstrip()
    complete_tag_list.append(BIO_tagged_string)

def B_and_I_tags(test_keys_list , test_text):
    B_tag_strings= []
    I_tag_strings=[]
    for i in range ( len(test_keys_list)):
        temp= test_keys_list[i][1:].split(" ")
        temp[0]=temp[0].replace("'" ,"")
        temp[0]=temp[0].replace("," ,"")
        B_tag_strings.append(temp[0])
        for j in range (len(temp)):
            if(j>0):
                I_tag_strings.append(temp[j])
    all_words = test_text.split()
    generate_BIO_tag(all_words , B_tag_strings , I_tag_strings)


def get_BIO_tags():
    for i in range(len(df)):
        text= df["text"][i]
        keys=df["keys"][i]
        keys_test= keys[1:len(keys)-1]
        keys_list = keys_test.split(",")
        B_and_I_tags(keys_list  , text)
       
        
        
```
