```python
stop_dir="./SmartStoplist.txt"
```

```python
rake_object= RAKE.Rake(stop_dir)
```

```python
def Sort_Tuple(tup):
  tup.sort(key= lambda x: x[1])
  return tup
```

```python
keywords= Sort_Tuple(rake_object.run(text))
```

```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/BIO_tagged_dataset_Modified2.csv')
```

```python
def get_keywords(keywords):
  temp= []
  for i in range(len(keywords)):
    temp.append(keywords[i][0])
  return temp
```

```python
for i in range (len(data)):
  keywords= Sort_Tuple(rake_object.run(data.iloc[i]['text']))
  list_keywords.append(str(get_keywords(keywords)))


```
